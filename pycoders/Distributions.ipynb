```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Parameters for different distributions
n, p = 20, 0.1  # Binomial distribution parameters: number of trials and probability of success
mu, sigma = 0, 10  # Normal distribution parameters: mean and standard deviation
lmbda = 5  # Poisson distribution parameter: average rate of occurrence
rate = 1 / 3  # Exponential distribution parameter: rate (inverse of mean)

# Generate random samples from different distributions
binomial_sample = np.random.binomial(n, p, 1000)
normal_sample = np.random.normal(mu, sigma, 1000)
poisson_sample = np.random.poisson(lmbda, 1000)
exponential_sample = np.random.exponential(rate, 1000)

# Create histograms of the random samples
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.hist(binomial_sample, bins=np.arange(n + 2) - 0.5, density=True, color='blue', alpha=0.7, edgecolor='k')
plt.title('Binomial Distribution (n = 10, p = 0.5)')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')

plt.subplot(2, 2, 2)
plt.hist(normal_sample, bins=30, density=True, color='green', alpha=0.7, edgecolor='k')
plt.title('Normal Distribution (μ = 0, σ = 1)')
plt.xlabel('Value')
plt.ylabel('Probability Density')

plt.subplot(2, 2, 3)
plt.hist(poisson_sample, bins=np.arange(poisson_sample.max() + 2) - 0.5, density=True, color='orange', alpha=0.7, edgecolor='k')
plt.title('Poisson Distribution (λ = 5)')
plt.xlabel('Number of Events')
plt.ylabel('Probability')

plt.subplot(2, 2, 4)
plt.hist(exponential_sample, bins=30, density=True, color='red', alpha=0.7, edgecolor='k')
plt.title('Exponential Distribution (rate = 1/3)')
plt.xlabel('Time Between Events')
plt.ylabel('Probability Density')

plt.tight_layout()
plt.show()
```
